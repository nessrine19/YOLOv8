##Training

```python
yolo = YOLO("yolov8n.pt")
yolo.train(data="/content/drive/MyDrive/Nesrine_YOLOv8/YOLO/yolo.yaml", epochs=10)
```

##Evaluate

```python
results = yolo.val()
print(results)
```

##Testing

```python
image_path = '/content/drive/MyDrive/Nesrine_YOLOv8/YOLO/Testing/7117_Caranx_sexfasciatus_juvenile_f001188.jpg'  # Change this to the path of the image you want to test

# Perform inference
results = yolo.predict(source=image_path)
```

```python
results
```

```python
import cv2

# Extract the original image and detections
orig_img = results[0].orig_img
boxes = results[0].boxes
print("number of fishes in this img : ",len(boxes))
# Draw bounding boxes and labels on the image
for box in boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
    label = results[0].names[int(box.cls[0])]  # Get class label
    confidence = box.conf[0]  # Get confidence score

    # Draw bounding box
    cv2.rectangle(orig_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Draw label and confidence
    label_text = f"{label}: {confidence:.2f}"
    cv2.putText(orig_img, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Save or display the resulting image
output_path = '/content/drive/MyDrive/Nesrine_YOLOv8/YOLO/OUTPUT/image.jpg'  # Change this to where you want to save the output image
cv2.imwrite(output_path, orig_img)
```

```python
#test on video
# Open the video file
input_video_path = '/content/drive/MyDrive/Nesrine_YOLOv8/YOLO/img_for_interference/fish_test1.mp4'  # Change this to the path of your input video
output_video_path = '/content/drive/MyDrive/Nesrine_YOLOv8/YOLO/OUTPUT/result.mp4'  # Change this to where you want to save the output video

cap = cv2.VideoCapture(input_video_path)

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')

# Create VideoWriter object to save the output video
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Perform inference on the frame
    results = yolo(frame)

    # Extract the detections
    boxes = results[0].boxes

    # Draw bounding boxes and labels on the frame
    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Get bounding box coordinates
        label = results[0].names[int(box.cls[0])]  # Get class label
        confidence = box.conf[0]  # Get confidence score

        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Draw label and confidence
        label_text = f"{label}: {confidence:.2f}"
        cv2.putText(frame, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Write the processed frame to the output video
    out.write(frame)

# Release the video objects
cap.release()
out.release()
cv2.destroyAllWindows()
```
